```python
import os
import warnings
warnings.filterwarnings('ignore')
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_huggingface import HuggingFaceEmbeddings

os.environ['GOOGLE_API_KEY'] = os.getenv('GOOGLE_API_KEY')
```

```python
loader = PyPDFLoader(r"E:\Learning\Agentic_AI_V2\assignment\1-RAG-PDF\data\agentic-ai-system.pdf")
pages = loader.load()
```

```python
pages = pages[12:]
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100, separators=['\n\n', '\n', ' ', ''])
splited_pages = splitter.split_documents(pages)
```

```python
len(pages)
```

```python
len(splited_pages)
```

```python
for doc in pages:
    print(len(doc.page_content))
```

```python
for doc in splited_pages:
    print(len(doc.page_content))
```

```python
embedding_hf = HuggingFaceEmbeddings(model='all-MiniLM-L6-v2')

text = [doc.page_content for doc in splited_pages]
vector_space_hf = embedding_hf.embed_documents(text)
```
